```python
import numpy as np
from nn import Net, LinearLayer, ReLu, SoftMax
from metrics import accuracy

from sklearn.datasets import load_digits
from sklearn import preprocessing
from sklearn.model_selection import train_test_split

data = load_digits()
x = data['data']
y = data['target']

x = preprocessing.normalize(x)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25)

dim_in = x_train.shape[1]
dim_out = len(np.unique(y_train))
dim_layer = 512

model = []
model.append(LinearLayer(dim_in, dim_layer, drop=0.2))
model.append(ReLu())
model.append(LinearLayer(dim_layer, int(dim_layer/2), drop=0.6))
model.append(ReLu())
model.append(LinearLayer(int(dim_layer/2), dim_out))
model.append(SoftMax())


net = Net(model, num_epoch=100, shuffle=True)
net.train(x_train, y_train, x_test, y_test, info=True)

y_pred = net.predict(x_test)
accuracy = accuracy(y_test, y_pred)
print(f'Test accuracy: {accuracy:.3f}')
```
