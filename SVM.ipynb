```python
from svm import SVC
import numpy as np
from sklearn import datasets
from sklearn import preprocessing
```

```python
data = datasets.load_iris()
#print(data.keys())
x = data['data']
y = data['target']
x = preprocessing.normalize(x)
```

```python
from metrics import accuracy
from kernels import gauss_kernel

clf = SVC(kernel=gauss_kernel, C=10, gamma=10)
clf.train(x, y)
y_pred = clf.predict(x)
accuracy(y, y_pred)
```

```python
from sklearn import svm
clf_sk = svm.SVC(probability=True, gamma=0.1)
clf_sk.fit(x, y)
y_pred = clf_sk.predict(x)
accuracy(y, y_pred)
```

```python
#VERY INTERESTING COMPARING WITH SKLEARN WE CAN SEE THAT THE DIFFERENCE BETWEEN CLASS 1 AND 2 IS MINIMUM
clf_sk.get_params()
clf_sk.predict_proba(x)
```
